```python
import pandas
import numpy as np

#Loads the data in a pandas Dataframes
 
train_df = pandas.read_csv("penguins.csv")
test_df = pandas.read_csv("penguins.csv")

#Display the fields

train_df.info()
```

```python
#Display how the data looks like
train_df.head()
```

```python
#name of the column 
classes = train_df["species"].unique().tolist()
print(f"Label classes: {classes}")

train_df["species"] = train_df["species"].map(classes.index)
```

```python
#Spliting the dataset into training and testing
def split_dataset(dataset, test_ratio=0.30):
  #Split a panda dataframe in two .
  test_indices = np.random.rand(len(dataset)) < test_ratio
  return dataset[~test_indices], dataset[test_indices]

train_df_pd, test_df_pd = split_dataset(train_df)
print("{} examples in training, {} examples for testing".format(len(train_df_pd),len(test_df_pd)))

```

```python
#Coverting the pandas dataframe into Tensorflow dataset
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train_df, label ="species")
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(test_df, label ="species")
```

```python
#Specify the model
model = tfdf.keras.RandomForestModel()

#train the model.
model.fit(train_ds)

#look at the model how it looks like
model.summary()

```

```python
#interpreting Model
tfdf.model_plotter.plot_model_in_colab(model, tree_idx=0)
```

```python
#evaluating the model
model.compile(metrics=["accuracy"])
score = model.evaluate(test_ds, return_dict=True)
print()

for name, value in score.items():
  print(f"{name}: {value:.4f}")
```
